## BARPLOT

```python
import matplotlib.pyplot as plt
import numpy as np
x=['A','B','C','D','E']
y=np.random.randint(low=1000,high=9000,size=5)
#figure sizd
plt.subplots(figsize=(8,4))
#create a bar chart
plt.bar(x,y,width=0.8)

#add lables and titles
plt.xlabel("Category")
plt.ylabel("Sales (in lakhs)")
plt.title("Category vs Sales")
plt.show()
```

```python
#horizontal bar chart
import matplotlib.pyplot as plt
import numpy as np
x=['A','B','C','D','E']
y=np.random.randint(low=1000,high=9000,size=5)
#figure size
plt.subplots(figsize=(8,5))
#create a bar chart
plt.barh(x,y,color=["red","blue","lightgreen","yellow","black"],height=0.7,edgecolor="darkblue")

#add lables and titles
plt.xlabel("Category")
plt.ylabel("Sales (in lakhs)")
plt.title("Category vs Sales")
plt.show()
```

```python
#horizontal bar chart
import matplotlib.pyplot as plt
import numpy as np
x=['A','B','C','D','E']
y=np.random.randint(low=1000,high=9000,size=5)
#figure size
plt.subplots(figsize=(8,5))
#create a bar chart
plt.bar(x,y,color=["red","blue","lightgreen","yellow","black"],edgecolor="black",linewidth=5)

#add lables and titles
plt.xlabel("Category")
plt.ylabel("Sales (in lakhs)")
plt.title("Category vs Sales")
plt.show()
```

```python
# set transparency
# usin alpha we can set transparency

category = ['A','B','C','D','E']

sales = np.random.randint(low = 1000,high = 9000,size=5)

#plotting data
plt.bar(category,sales,
alpha=0.25,
color=['red','blue','green','orange','yellow'],
edgecolor='black',
linewidth=3) #alpha lies from 0 to 1, alpha is transparency

plt.xlabel('Category')
plt.ylabel('Sales (in lakhs)')
plt.title('Category vs Sales')
plt.show()
```

## Plotting Pie Chart

USING ADDITIONAL FEATURES : COLORS,EXPLODE,SHADOW,STARTANGLE

```python
import matplotlib.pyplot as plt
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # only "explode" the 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Basic Pie Chart')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['pink','purple','red','skyblue']
explode = (0.1, 0, 0, 0)

fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},
    textprops={'fontsize': 14, 'fontweight': 'bold'}
)


ax.axis('equal')

plt.title('Pie Chart')


plt.legend(wedges, labels, title="Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12)

plt.show()
```

### Adding grid and annotations

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.plot(x, y, color='purple', marker='s', linestyle='-', linewidth=2, label='y = x^2')

plt.title('Line Chart with Grid and Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.grid(True)

plt.annotate('Local max', xy=(4, 16), xytext=(3, 20),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt


x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [0, 1, 8, 27, 64, 125]


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='pink', marker='o', linestyle='--', linewidth=2, label='y = x^2')
plt.title('Subplot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(x, y2, color='purple', marker='x', linestyle='-', linewidth=2, label='y = x^3')
plt.title('Subplot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()


plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,4,9,16,25]
y3 = [0, 1, 8, 27, 64]
y4 = [25, 20, 15, 10, 5]
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
# First subplot
axs[0, 0].plot(x, y1, color='blue', marker='o', linestyle='--', linewidth=2)
axs[0, 0].set_title('Subplot 1')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].legend()

# Second subplot
axs[0, 1].plot(x, y2, color='green', marker='x', linestyle='-', linewidth=2)
axs[0, 1].set_title('Subplot 2')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].legend()

# Third subplot
axs[1, 0].plot(x, y3, color='red', marker='s', linestyle='-.', linewidth=2)
axs[1, 0].set_title('Subplot 3')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].legend()
# Fourth subplot
axs[1, 1].plot(x, y4, color='purple', marker='d', linestyle=':', linewidth=2)
axs[1, 1].set_title('Subplot 4')
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

## Scatter plot

```python


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [20000, 30000, 45000, 35000, 28000, 50000]

plt.figure(figsize=(8, 5))
plt.scatter(months, sales, color='blue', marker='o', s=100, alpha=0.75, edgecolors='black')

plt.title('Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales ($)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(42)
scores = np.random.normal(loc=60, scale=10, size=100)  # generating 100 scores with mean 60 and standard deviation 10

plt.figure(figsize=(8, 6))
plt.hist(scores, bins=10, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=70, scale=15, size=100)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))


axs[0].hist(data1, bins=15, color='purple', edgecolor='black', alpha=0.7)
axs[0].set_title('Histogram of Data 1')
axs[0].set_xlabel('Values')
axs[0].set_ylabel('Frequency')
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

axs[1].hist(data2, bins=15, color='pink', edgecolor='black', alpha=0.7)
axs[1].set_title('Histogram of Data 2')
axs[1].set_xlabel('Values')
axs[1].set_ylabel('Frequency')
axs[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```

## Area plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y1 = np.sin(x)                # sine curve
y2 = np.cos(x)                # cosine curve


plt.figure(figsize=(8, 6))

plt.fill_between(x, y1, color="pink", alpha=0.4, label='Sine Curve Area')
plt.fill_between(x, y2, color="purple", alpha=0.4, label='Cosine Curve Area')

plt.plot(x, y1, color="pink", alpha=0.6, linewidth=2, label='Sine Curve')
plt.plot(x, y2, color="purple", alpha=0.6, linewidth=2, label='Cosine Curve')


plt.title('Area Plot of Sine and Cosine Curves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```

## Box plot

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(110, 5, 200)
data = [data1, data2, data3]

plt.figure(figsize=(10, 6))
plt.boxplot(data, patch_artist=True, notch=True, vert=True, showmeans=True)

plt.title('Box Plot of Sample Data')
plt.xlabel('Group')
plt.ylabel('Values')
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])


plt.show()
```
