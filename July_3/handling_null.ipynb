## Imputing Null Values
**fillna()**: Fills the null values with a specific value or a method

```python
data = df2["Diastolic_BP"].fillna(value = 0) # it will replace all null values with 0
data.head()

data.isnull().sum() # Checks if there is any null values present
```

## Handling Missing Data

'pad' method is used for replacing the NaN values with the value of the row above it.
Here , the first three NaN values have null. So there is no previous row which has a value for it . The data still contains NaN values

```python
data1 = df2.fillna(method = 'pad') # Nan values are replaced with the value of the row above i
data1
data1.isnull().sum()
```

We can fix this with using 'bfill' method which takes the value of the row below it .

```python
data2 = data1.fillna(method = 'bfill')
data2
data2.isnull().sum()
```

## Use Mean, Median, Mode, Max, and Min

```python
import matplotlib.pyplot as plt
from scipy.stats import skew
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
axes = axes.flatten()

for i, col in enumerate(df2.columns):
    sns.histplot(df2[col], bins=10, kde=True, ax=axes[i])
    axes[i].set_title(f'{col} (Skewness: {skew(df2[col]):.2f})')

plt.tight_layout()
plt.show()
```

Some of the graphs have skew distribution because they have outliers thus we will u

let's look at the null values of each coloumn and decide which method should we use to replace the null values with.

```python
df2.isnull().sum()
```

We will use mean when we have numerical data with normal distribution. Mean is used when data has few outliers. So for 'Glucose' coloumn I have replaced the null values with the mean of the coloumn

```python
val = df2['Glucose'].mean()
df4 = df2['Glucose'].fillna(value = val)
df4.isnull().sum()
df4
```

"Skin_Fold" has skewed data and outliers are present. Hence we used median to replace the null values

```python
val = df2['Skin_Fold'].median()
df5 = df2['Skin_Fold'].fillna(value = val)
df5.isnull().sum()
df5
```

```python
a=sns.histplot(df4, bins=10, kde=True)
a.set_title("Glucose after repalcing the null values with mean")
```

```python
b=sns.histplot(df5, bins=10, kde=True)
b.set_title("Skin_fold after repalcing the null values with median")
```

## Interpolation

```python
df_interpolated = df2.copy()
df_interpolated['Serum_Insulin']= df2['Serum_Insulin'].interpolate(method='linear')

plt.figure(figsize=(18,8))
plt.subplot(1,2,1)
sns.histplot(df2['Serum_Insulin'],kde = True, color = 'pink', bins = 40) # Original
plt.title("Original Data Distribution")

plt.subplot(1,2,2)
sns.histplot(df_interpolated['Serum_Insulin'],kde = True, color = 'skyblue', bins = 40) # Interpolated
plt.title("Data Distribution after interpolation")

plt.tight_layout()
```

Interpolate missing values using polynomial interpolation of degree 2.
Polynomial interpolation is useful when the relationship between the data points is linear

```python
df9 = df2.interpolate(method='polynomial', order=2)
df9
```

Interpolate missing categorical values using the most recent non-null value.

## Imputation Techniques

KNN IMPUTATION

```python
from sklearn.impute import KNNImputer
columns_for_knn = ['Diastolic_BP', 'Skin_Fold', 'Serum_Insulin']

# Perform KNN imputation
imputer = KNNImputer(n_neighbors=5)
df_knn_imputed = pd.DataFrame(imputer.fit_transform(df2[columns_for_knn]), columns=columns_for_knn)

# Visualize histograms after imputation
plt.figure(figsize=(14, 10))
for i, col in enumerate(columns_for_knn, 1):
    plt.subplot(3, 2, i)
    plt.hist(df_knn_imputed[col], bins=30, color='purple', edgecolor='black')
    plt.title(f'Histogram of {col} - After KNN Imputation')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

MICE (Multiple Imputation by Chained Equations)

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
columns_for_knn = ['Diastolic_BP', 'Skin_Fold', 'Serum_Insulin']

# Perform MICE imputation
imputer_mice = IterativeImputer(random_state=0)
df_mice_imputed = pd.DataFrame(imputer_mice.fit_transform(df2[columns_for_knn]), columns=columns_for_knn)

# Visualize histograms after imputation
plt.figure(figsize=(14, 10))
for i, col in enumerate(columns_for_knn, 1):
    plt.subplot(3, 2, i)
    plt.hist(df_mice_imputed[col], bins=30, color='orange', edgecolor='black')
    plt.title(f'Histogram of {col} - After MICE Imputation')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
