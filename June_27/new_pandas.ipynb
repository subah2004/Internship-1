## Series
one dimensional arrays that can hold any datatypes

```python
import pandas as pd

data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print(series)
```

## DataFrame

A two-dimensional labeled data structure with columns that can be of different types

```python
data = {
    'Name': ['Subah', 'Alex', 'Sonakashi'],
    'Age': [20, 30, 35],
    'City': ['Dharamshala', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
print(df)
```


## Exploring Series Attributes and Methods

```python
print("Index:", series.index)
print("Values:", series.values)
print("Data type:", series.dtype)
print("Size:", series.size)
```

```python
print("Mean of the Series:", series.mean())
print("sorted values:", series.sort_values())
print("First two elements:", series.head(2))
print("Last two elements:", series.tail(2))
```
