## Line Plot


```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Line Plot')

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot')

plt.grid(True)
plt.legend()
plt.show()
```

Using dataset


```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")

plt.plot(data.index, data['total_bill'])

plt.xlabel('Index')
plt.ylabel('Total Bill ($)')
plt.title('Total Bill in Tips Dataset')

plt.grid(True)

plt.legend()

plt.show()
```

```python

flights_data = sns.load_dataset("flights")

plt.figure(figsize=(10, 6))
sns.lineplot(data=flights_data, x='year', y='passengers', marker='o', color='g')

plt.xlabel('Year')
plt.ylabel('Passengers')
plt.title('Passenger Traffic Over Time')

plt.grid(True)

plt.show()
```

## Histogram


```python

iris_data = sns.load_dataset("iris")

setosa_data = iris_data[iris_data['species'] == 'setosa']


plt.figure(figsize=(8, 6))
sns.histplot(setosa_data['petal_width'], bins=10, kde=True, color='pink')

plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Widths for Setosa Iris')


plt.show()
```

using multiple


```python

tips_data = sns.load_dataset("tips")
plt.figure(figsize=(10, 6))
sns.histplot(data=tips_data, x='total_bill', bins=20, kde=True, hue='sex', multiple='stack', palette='Set2')

plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bill Amounts by Gender')


plt.legend(title='Gender')

plt.show()
```

using element parameter



```python
sns.histplot(data=tips_data, x='total_bill', bins=20, kde=True, hue='sex', palette='Set2',element='step')
plt.show()
```

## Bar plot

```python
titanic_data = sns.load_dataset("titanic")

survival_count = titanic_data.groupby('sex')['survived'].sum().reset_index()

sns.barplot(data=survival_count, x='sex', y='survived', palette='Set2')

plt.xlabel('Sex')
plt.ylabel('Number of Survivors')
plt.title('Number of Titanic Passengers Who Survived by Sex')

plt.show()
```

```python
print("Original tips dataset:")
print(tips_data.head())

tips_wide = tips_data.pivot_table(index='day', columns='sex', values='total_bill', aggfunc='sum')

print("Wide format tips dataset:")
print(tips_wide)
```

```python
sns.barplot(tips_wide)
```

## HeatMap


```python
flights_wide = flights_data.pivot(index='year', columns='month', values='passengers')

plt.figure(figsize=(10, 8))
sns.heatmap(flights_wide, annot=True, fmt='d', cmap='YlGnBu')
```
